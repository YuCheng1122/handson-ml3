加載資料

```python
import pandas as pd
import glob

# 設定你的目錄路徑
directory_path = 'C:/Users/Tommy/Desktop/eventData/AICTD/'

# 使用glob來找到該目錄下的所有CSV檔案
csv_files = glob.glob(directory_path + '*.csv')

# 用於儲存所有DataFrame的列表
dfs = []

# 遍歷找到的CSV檔案，並將每個檔案讀取為一個DataFrame
for file in csv_files:
    df = pd.read_csv(file)
    dfs.append(df)

# 合併所有DataFrame
combined_df_train = pd.concat(dfs, ignore_index=True)
combined_df_train
```

檢查缺失值

```python
combined_df_train.isnull().sum()
```

清除 Metadata這一欄

```python
combined_df_train.drop('Metadata', axis=1, inplace=True)
```

資料前處理 - 時間欄位

```python
import pandas as pd

# 假設df是您的DataFrame，並且已經包含了'timestamp'列
# 轉換 timestamp 為 datetime 對象，並提取小時和分鐘
combined_df_train['timestamp'] = pd.to_datetime(combined_df_train['Timestamp'])
combined_df_train['hour'] = combined_df_train['timestamp'].dt.hour
combined_df_train['minute'] = combined_df_train['timestamp'].dt.minute

# 定義一個函數來判斷時間是否為上班時間


def is_work_time(hour):
    return '上班' if 9 <= hour < 18 else '下班'


# 使用apply函數應用上述函數
combined_df_train['work_time'] = combined_df_train['hour'].apply(is_work_time)

combined_df_train
```

資料前處理 - 類別資料

```python
from sklearn.preprocessing import LabelEncoder

# 為 category 和 signature 創建獨立的 LabelEncoder 實例
signature_encoder = LabelEncoder()
category_encoder = LabelEncoder()

combined_df_train['category'] = category_encoder.fit_transform(
    combined_df_train['Category'])
combined_df_train['signature'] = signature_encoder.fit_transform(
    combined_df_train['Signature'])

# 顯示 signature
print("Signature classes:", list(signature_encoder.classes_))
print("Signature encoded values:",
      signature_encoder.transform(signature_encoder.classes_))

# 顯示 Category
print("Category classes:", list(category_encoder.classes_))
print("Category encoded values:",
      category_encoder.transform(category_encoder.classes_))

combined_df_train
```

資料前處理 - IP

```python
import pandas as pd
import ipaddress


def ip_to_bytes(ip):
    try:
        ip_obj = ipaddress.ip_address(ip)
        return [int(byte) for byte in ip_obj.packed]
    except ValueError:
        return [None, None, None, None]


# 假設 combined_df 是你的 DataFrame
combined_df_train['Src_IP_Byte1'], combined_df_train['Src_IP_Byte2'], combined_df_train['Src_IP_Byte3'], combined_df_train['Src_IP_Byte4'] = zip(
    *combined_df_train['Src IP'].apply(ip_to_bytes))
combined_df_train['Dest_IP_Byte1'], combined_df_train['Dest_IP_Byte2'], combined_df_train['Dest_IP_Byte3'], combined_df_train['Dest_IP_Byte4'] = zip(
    *combined_df_train['Dest IP'].apply(ip_to_bytes))
combined_df_train
```

數值特徵標準化

```python
from sklearn.preprocessing import StandardScaler

# 選擇需要標準化的數值特徵
numeric_features = ['hour', 'minute', 'category', 'signature', 'Src Port', 'Dest Port', 'Src_IP_Byte1', 'Src_IP_Byte2', 'Src_IP_Byte3',
                    'Src_IP_Byte4', 'Dest_IP_Byte1', 'Dest_IP_Byte2', 'Dest_IP_Byte3', 'Dest_IP_Byte4']

# 初始化 StandardScaler
scaler = StandardScaler()

# 對數值特徵進行標準化
combined_df_train[numeric_features] = scaler.fit_transform(
    combined_df_train[numeric_features])
```

```python
combined_df_train
```

nu 最佳值

```python
from sklearn.svm import OneClassSVM
from sklearn.model_selection import GridSearchCV
import numpy as np

# 定義特徵
features = ['hour', 'minute', 'category', 'signature', 'Src Port', 'Dest Port', 'Src_IP_Byte1', 'Src_IP_Byte2', 'Src_IP_Byte3',
                    'Src_IP_Byte4', 'Dest_IP_Byte1', 'Dest_IP_Byte2', 'Dest_IP_Byte3', 'Dest_IP_Byte4']
X_train = combined_df_train[features]
# 設置 nu 值的範圍
nu_range = np.linspace(0.01, 0.1, 10)

# 創建 OneClassSVM 模型
ocsvm = OneClassSVM(kernel='rbf', gamma='scale')

# 使用 GridSearchCV 來尋找最佳 nu
param_grid = {'nu': nu_range}
grid_search = GridSearchCV(ocsvm, param_grid, cv=5)  # cv 是交叉驗證的次數

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)
grid_search.fit(X_scaled)

# 最佳 nu 值
best_nu = grid_search.best_params_['nu']
print("Best nu:", best_nu)
```

訓練OCSVM模型

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import OneClassSVM

# 定義特徵
features = ['hour', 'minute', 'category', 'signature', 'Src Port', 'Dest Port', 'Src_IP_Byte1', 'Src_IP_Byte2', 'Src_IP_Byte3',
                    'Src_IP_Byte4', 'Dest_IP_Byte1', 'Dest_IP_Byte2', 'Dest_IP_Byte3', 'Dest_IP_Byte4']
X_train = combined_df_train[features]

# 規範化或標準化數據
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)

# 創建 One-Class SVM 模型
# 調整 nu 參數和核函數
model = OneClassSVM(nu=0.1, kernel='rbf', gamma='scale')
model.fit(X_scaled)  # 使用標準化後的數據來訓練模型
```

模型保存

```python
import pickle

# 保存模型
with open('oneclass_svm_model_v5.pkl', 'wb') as file:
    pickle.dump(model, file)
```
